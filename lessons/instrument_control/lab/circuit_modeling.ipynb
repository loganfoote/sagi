```python
import numpy as np
import matplotlib.pyplot as plt
```

# Circuit Modeling

```python
# - frequency vector - #
f = np.linspace(1e3, 500e3, int(1e5))

# - lowpass filter - #
Rlp = 330
Clp = 10e-9
Hlp = 1 / (1 + 2*np.pi*f*Rlp*Clp*1j)

# - highpass filter - #
Rhp = 169
Chp = 100e-9
Hhp = 2*np.pi*f*Rhp*Chp*1j / (1 + 2*np.pi*f*Rhp*Chp*1j)

# - bandpass filter - #
Hbp = Hlp * Hhp

# - plot - #
fig, ax = plt.subplots()
ax.set(
    xscale='log',
    xlabel='Frequency (Hz.)',
    ylabel='(dB)',
    title='Filter Simulation'
)
ax.plot(f, 20*np.log10(np.abs(Hbp)), label='Band-pass magnitude response')
ax.plot(f, 20*np.log10(np.abs(Hlp)), label='Low pass magnitude response', linestyle=':')
ax.plot(f, 20*np.log10(np.abs(Hhp)), label='High pass magnitude response', linestyle=':')

ax.legend()

fig.tight_layout()

fig.savefig('figures/20240731_BandpassSimulation.png', dpi=300)
```
